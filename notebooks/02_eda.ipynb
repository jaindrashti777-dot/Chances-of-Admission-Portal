# Exploratory Data Analysis (EDA)

In this notebook, we perform a deep dive into the dataset features and how they correlate with `Admission_Chance`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent))
from src.data_loader import load_raw_data
from src.utils import ensure_dir

sns.set_theme(style="whitegrid")
FIGURES_DIR = Path("../reports/figures")
ensure_dir(FIGURES_DIR)

df = load_raw_data()
```

## 1. Target Variable Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Admission_Chance'], bins=50, kde=True, color="#2196F3")
plt.title("Distribution of Admission Chance", fontsize=14, fontweight="bold")
plt.xlabel("Admission Chance")
plt.ylabel("Frequency")
plt.savefig(FIGURES_DIR / "target_distribution.png", dpi=150)
plt.show()
```

## 2. Univariate Analysis (Academic Scores)

```python
academic_cols = ['Tenth_Percentage', 'Twelfth_Percentage', 'JEE_Percentile', 'CUET_Score', 'CGPA']
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, col in enumerate(academic_cols):
    sns.histplot(df[col], bins=30, kde=True, ax=axes[i], color="#4CAF50")
    axes[i].set_title(f"Distribution of {col}", fontweight="bold")

fig.delaxes(axes[-1])
plt.tight_layout()
plt.show()
```

## 3. Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="CGPA", y="Admission_Chance", hue="College_Tier", alpha=0.6)
plt.title("CGPA vs Admission Chance (Colored by College Tier)", fontsize=14, fontweight="bold")
plt.savefig(FIGURES_DIR / "cgpa_vs_admission.png", dpi=150)
plt.show()
```

## 4. Correlation Heatmap

```python
plt.figure(figsize=(12, 10))
num_cols = df.select_dtypes(include=[np.number]).columns
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", vmin=-1, vmax=1)
plt.title("Correlation Heatmap of Numerical Features", fontsize=15, fontweight="bold")
plt.savefig(FIGURES_DIR / "correlation_heatmap.png", dpi=150)
plt.show()
```

### Q&A
- **What are the highest correlations with Admission_Chance?** Usually CGPA, JEE Percentile, and 12th Percentage show strong positive correlation.
- **Does College Tier matter?** Yes, from the scatterplot, Tier 1 students might have different baselines.

### Data Analysis Key Findings
- Strong correlation between all academic scores, as expected.
- Backlogs have a negative correlation with Admission Chance.
